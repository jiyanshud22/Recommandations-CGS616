```python
df = pd.read_csv('u.data', sep='\t', encoding='ISO-8859-1', header=None,names=['user id' , 'item id' , 'rating' ,'timestamp'])
pivot_table = df.pivot_table(index='user id', columns='item id', values='rating', aggfunc='first')
# Fill missing values with empty strings
pivot_table = pivot_table.fillna(0)
new_data=pd.DataFrame(pivot_table)
new_data
```

```python
predicted_ratind_df=pd.read_csv('predicted_ratings_df.csv')
predicted_ratind_df
```

```python
user_id=int(input("Enter User id:"))

# Calculate cosine similarity between users
similarity=cosine_similarity(predicted_ratind_df)

# Find most similar user
x=list(sorted(enumerate(similarity[user_id-1]), key=lambda x:x[1],reverse=True)[1:2])
#print(x)

# Get items rated by both users
user1_items = set(df[df['user id'] == user_id]['item id']) 
user2_items = set(df[df['user id'] == x[0][0]+1]['item id'])

print('Recommend Movie:')

# Find movies recommended to similar user but not rated by the current user
intersection_movies = user2_items-user1_items
#print(intersection_movies)
movie=list(intersection_movies)
count=0
#print(movie)
# Extract ratings for recommended movies
rating=[]
for i in movie:
    rating.append(df[df['item id'] == i]["rating"].values[0])

# Combine ratings with movie IDs
new_list = [(x, y) for x, y in zip(rating, movie)]
movies_rating=sorted(new_list,reverse=True)
for i in movies_rating:
    movie_title=item[item['movie id']==i[1]]['movie title'].values[0]
    print(movie_title,i[0])
    if count<4:# Print only top 5 recommendations
        count+=1
    else:
        break
```
